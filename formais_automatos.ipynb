```python
class Pedidos:
    def __init__(self,lanche,pagamento, valor):
        self.lanche = lanche
        self.pagamento = pagamento
        self.valor= valor
```

```python
class Realizar_Pedidos(Pedidos):
    def __init__ (self,lanche,pagamento,valor):
        super().__init__(lanche,pagamento,valor)
    
    def cancelar_pedido(self, valor):
        self.valor = valor
        if int(valor) < 0:
            print("Pedido cancelado Por falta de credito")
            finalizado(1)
        else:
            print('Pedido sendo preparado')
    def cancelar(self, cod):
        self.cod = cod
        if cod == 1:
            print("Pedido Cancelado")
            self.finalizado(cod)
        else:
            print("Carregando...")
        
    def pedido_sendo_preparado(self,cod):
        self.cod = cod
        if cod == 1:
            self.cancelar(cod)
        else:
            print("Pedido esta sendo preparado")
    def pedido_saiu_entrega(self,cod):
        self.cod =  cod
        if cod == 1:
            self.cancelar(cod)
        else:
            print("Seu pedido saiu para entregue")
    def pedido_entregue(self,cod):
        if cod == 0:
            print("Seu pedido foi entregue!!!")
            self.finalizado(1)
        else:
            self.cancelar(cod)
            
    def finalizado(self,cod):
        self.cod = cod
        if cod == 1:
            print("Pedido Finalizado")            
        
```

### Questão 03

#### Item A:
* Cadeias aceitas : {kok, okok, kkkoookkk}
* Cadeias não aceitas :{kkkkk, ko, koko, ooo, koo} 

```python
import re
def okok(cadeia):
    if (re.search(r'ok',cadeia)):
        return print("Cadeia aceita")
    else:
        return print('Não aceita')
```

```python
okok('okoko')
```
